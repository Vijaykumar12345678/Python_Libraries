1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
def plot(xvalues,yvalues):
    """
    Description:
        This function will plot the line chart
    
    parameters:
        xvalues : ndarray consisting of data
        yvalues : ndarray consisting of data
    
    returns:
        None
    """
    plt.figure(figsize=(10,6))
    plt.xlabel("Months")
    plt.ylabel("Sales")
    plt.title("Sales Over Time")
    plt.plot(xvalues,yvalues)
    plt.show()

def main():

    xvalues=np.array([1,2,3,4,5])
    yvalues=np.array([10,20,30,40,50])
    plot(xvalues,yvalues)
if __name__=="__main__":
    main()
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line.

```python

def plot_multiple_lines(data):
    """
    Description:
        This function is used to multiple line on a single plot
    
    parameters:
        data:Its consisting of data which has to be plot
        
    Returns:
        None
    """
    plt.figure(figsize=(10, 6))
    
    for line_data in data:
        plt.plot(line_data['x'], line_data['y'], label=line_data['label'], marker='o')
    
    plt.xlabel('X Axis Label')
    plt.ylabel('Y Axis Label')
    plt.title('Multiple Lines on Same Plot')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
   
    data = [
        {'x': [1, 2, 3, 4, 5], 'y': [1, 4, 9, 16, 25], 'label': 'Square'},
        {'x': [1, 2, 3, 4, 5], 'y': [1, 8, 27, 64, 125], 'label': 'Cube'},
        {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10], 'label': 'Linear'},
    ]
    
    plot_multiple_lines(data)

if __name__ == "__main__":
    main()
```

```python
def plot_multiple_lines(data):
    """
    Description:
        This function is used to multiple line on a single plot with legends and different line styles.
    
    parameters:
        data:Its consisting of data which has to be plot
        
    Returns:
        None
    """
    plt.figure(figsize=(10, 6))
    
    for line_data in data:
        plt.plot(line_data['x'], line_data['y'], label=line_data['label'], ls=line_data['ls'],marker='o')
        
    
    plt.xlabel('X Axis Label')
    plt.ylabel('Y Axis Label')
    plt.title('Multiple Lines on Same Plot')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
   
    data = [
        {'x': [1, 2, 3, 4, 5], 'y': [1, 4, 9, 16, 25], 'label': 'Square','ls':'solid'},
        {'x': [1, 2, 3, 4, 5], 'y': [1, 8, 27, 64, 125], 'label': 'Cube','ls':'dotted'},
        {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10], 'label': 'Linear','ls':'dashdot'},
    ]
    
    plot_multiple_lines(data)

if __name__ == "__main__":
    main()

    
```

9. Write a Python program to display the current axis limits values and set new axis values.


```python
import matplotlib.pyplot as plt

def display_and_set_axis_limits(xvalues,yvalues):
   
    """
    Description:
        This function takes in two lists of values, xvalues and yvalues, and plots them using
    
    parameters:
        xvalues (list): A list of x values to be plotted.
        yvalues (list): A list of y values to be plotted.
    
    Returns:
        None

    """
    plt.plot(xvalues, yvalues)
    # Display the current axis limits
    current_limits = plt.axis()
    print(f"Current axis limits: {current_limits}")
    
    # Set new axis limits
    plt.xlim(0, 6)  
    plt.ylim(0, 30)  
    
    # Display the new axis limits
    new_limits = plt.axis()
    print(f"New axis limits: {new_limits}")
    
    
    plt.title("Axis Limits Example")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

def main():
     # Sample data
    xvalues= [1, 2, 3, 4, 5]
    yvalues = [1, 4, 9, 16, 25]
    
    display_and_set_axis_limits(xvalues,yvalues)

if __name__ == "__main__":
    main()
```

12. Write a Python program to create multiple types of charts

```python

def create_charts_with_subplots():

    """
    Description:
        This is used to implement the concept of subplots.
    
    Parameters:
        None
    
    Returns:
        None"""
    # Data for the charts
    x_positions = [1, 2, 3, 4, 5]
    y_quantities = [10, 15, 20, 25, 30]
    
    categories = ['A', 'B', 'C', 'D']
    values = [23, 17, 35, 29]
    
    pie_labels = ['Category A', 'Category B', 'Category C', 'Category D']
    pie_sizes = [23, 17, 35, 25]
    
    # Create a figure with subplots
    plt.figure(figsize=(10, 6))
    
    # Line Chart
    plt.subplot(1,3,1)
    plt.plot(x_positions, y_quantities, marker='o', linestyle='-', color='blue')
    plt.title("Line Chart")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    
    # Bar Chart
    plt.subplot(1,3,2)
    plt.xlabel("Categories")
    plt.title("Bar Chart")
    plt.bar(categories, values, color='green')
    plt.ylabel("Values")
    
    # Pie Chart
    plt.subplot(1,3,3)
    plt.pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
    plt.title("Pie Chart")
    
   
    # Adjust layout to prevent overlap
    plt.tight_layout()
    
    # Display the plots
    plt.show()

def main():
    create_charts_with_subplots()

if __name__ == "__main__":
    main()
```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

```python
def plot_alphabet_closing_values(df):
    """
    Description:
            This function is used to  display the grid and draw line charts and  Customized the grid lines with
            rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
            off ticks.
            
    Parameters:
        df: dataframe consisiting of data
    
    
    Returns:
        None"""
    
    plt.figure(figsize=(10,6))

    
    plt.plot(df['Date'], df['Close'], marker='o', color='blue', linestyle='-')
    
    
    plt.grid(True, which='major', color='gray', linestyle='-', linewidth=0.8) 
    plt.minorticks_on()
    plt.grid(True, which='minor', color='lightgray', linestyle='--', linewidth=0.5)  
    
    # Turn off the ticks
    plt.tick_params(which='both', bottom=False, top=False, left=False, right=False, labelbottom=True)
    
    
    plt.title("Alphabet Inc. Closing Values (Oct 3, 2016 - Oct 7, 2016)")
    plt.xlabel("Date")
    plt.ylabel("Closing Value (USD)")
    
    
    plt.show()

def main():
     df =pd.read_csv("Sample_Finicial_Data.csv", parse_dates=['Date'], dayfirst=True)
     plot_alphabet_closing_values(df)

if __name__ == "__main__":
    main()
```
