1. Write a Python programming to create a pie chart of the popularity of programming
Languages.


```python
def pie_chart(programming_languages,popularity):
    """
    Description:
        This function generates a pie chart to display the popularity of different programming languages.

    Parameters:
        programming_languages: list  of programming languages
        popularity: list of their respective popularity

    Returns: 
        None
    """
    plt.figure(figsize=(8,10))
    plt.pie(popularity,labels=programming_languages,explode=[0,0.3,0,0,0,0],autopct='%2.1f%%',shadow=True)
    plt.legend()
    plt.show()

def main():
    Programming_languages=[ 'Java',' Python', 'PHP', 'JavaScript', 'C#', 'C++']
    Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    pie_chart(Programming_languages,Popularity)
if __name__=="__main__":
    main()
```

2. Write a Python programming to create a pie chart with a title of the popularity of programming
Languages.

```python
def pie_chart(programming_languages,popularity):
    """
    Description:
        This function generates a pie chart to display the popularity of different programming languages and along with the title

    Parameters:
        programming_languages: list  of programming languages
        popularity: list of their respective popularity

    Returns: 
        None
    """
    plt.figure(figsize=(8,10))
    plt.pie(popularity,labels=programming_languages,explode=[0,0.3,0,0,0,0],autopct='%2.1f%%',shadow=True)
    plt.title("Popularity of the programming languages")
    plt.legend()
    plt.show()

def main():
    Programming_languages=[ 'Java',' Python', 'PHP', 'JavaScript', 'C#', 'C++']
    Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    pie_chart(Programming_languages,Popularity)
if __name__=="__main__":
    main()
```

3. Write a Python programming to create a pie chart of gold medal achievements of five most
successful countries in 2016 Summer Olympics. Read the data from a csv file.


```python
def pie_chart(df):
    """
    Description:
        This function is to the plot the pie chart for the countries who won the top 5 countries medals in the olympics
        
    Parameters:
        df:dataFrame is consisting of rows and columns
    
    Returns:
        None"""
    plt.figure(figsize=(8,10))
    plt.pie(df['gold_medal'],labels=df['country'],shadow=True,autopct='%2.1f%%',explode=[0,0.2,0,0])
    plt.legend(title="countries")
    plt.show()

def main():
    df=pd.read_csv('medal.csv',delimiter=',')
    pie_chart(df)

if __name__=="__main__":
    main()
```
