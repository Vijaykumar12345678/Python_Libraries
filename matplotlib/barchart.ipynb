1.Write a Python programming to display a bar chart of the popularity of programming
Languages.

```python
def plot(programming_languages,popularity):
    """
    Description:
        This function is used to plot the bar chart for the popularity of the programming languages
        
    Parameters:
        programming_languages: list contaning the programming languages
        popularity: list continaing the the popularity of the programming languages
        
    Returns:
        None"""
    plt.figure(figsize=(10,6))
    colors=['red','black','blue','green','cyan','grey']
    plt.bar(programming_languages,popularity,color=colors,width=0.5)
    plt.xlabel("programming languages")
    plt.ylabel("popularity")
    plt.title("programming  languages popularity")

    plt.show()


def main():
    Programming_languages=[ 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    Popularity=[ 22.2, 17.6, 8.8, 8, 7.7, 6.7]
    plot(Programming_languages,Popularity)
if __name__=="__main__":
    main()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.

```python
def  plot_data(men_means,women_means):
    """
    Description:
        This function is used to plot the graph where grouping the gender and Use multiple X
        values on the same chart.
    
    Parameters:
        men_means: The means of the men.
        women_means:The Means of the women.
    
    Returns:  
        None """


    
    plt.subplots()
    index = np.arange(len(men_means))
    bar_width = 0.35
  

    plt.bar(index, men_means, bar_width,color='b',label='Men')

    plt.bar(index + bar_width, women_means, bar_width,color='g',label='Women')

    plt.xlabel('Person')
    plt.ylabel('Scores')
    plt.title('Scores by person')
    plt.xticks(index + bar_width/2, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
    plt.legend()

    plt.tight_layout()
    plt.show()
def main():
    men_means = (22, 30, 33, 30, 26)
    women_means = (25, 32, 30, 35, 29)
    plot_data(men_means,women_means)
if __name__=="__main__":
    main()
```

```python
def create_dataframe(df):
    """
    Description:
         This function is used to create a dataframe.
    
    Parameters:
        df: The data which has to be converted to dataframe

    Returns:
        Dataframe
        """
    return  pd.DataFrame(df,columns=['a','b','c','d','e'],index=[2,4,6,8,10])
     
    
def plot(df):
    """
    Description:
        This function is used to plot the bar chart.
        
    Parameters:
        df:dataframe
    
    Returns:
        None"""
    df.plot(kind='bar')


def main():
    array=np.array([[4,8,5,7,6],[2,3,4,2,6],[4,7,4,7,8],[2,6,4,8,6],[2,4,3,3,2]])
    df=create_dataframe(array)
    plot(df)
   
if __name__=="__main__":
    main()
```

12. Write a Python program to create bar plots with error bars on the same figure

```python
import numpy as np
import matplotlib.pyplot as plt

def plot( mean_velocity, standard_velocity):
    """ 
    Description:
        This function is used to plot the  mean velocity and standard velocity of the particles in the simulation.

        
    Parameters:
        mean_velocity: its consisting of the  mean velocity of the particles in the simulation.
        standard_velocity: its consisting of the standard velocity of the particles in the simulation.  
        
    Returns:
        None
    """
    plt.figure(figsize=(10, 6))
    positions=np.arange(len(mean_velocity))
   
    bars = plt.bar(positions, mean_velocity, yerr=standard_velocity,capsize=5,color='r', width=0.3)
    
   
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height, f'{height:.2f}', ha='center', va='bottom')
    
    
    plt.xlabel(" mean Velocity")
    plt.ylabel("Scores")
    plt.title("Scores by Velocity")
    
    plt.xticks(np.arange(len(mean_velocity)), positions)
    
   
    plt.show()

def main():
    
    
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    standard_velocity = [0.3314, 0.2278, 0.2836, 0.2645]  
    
    
    plot( mean_velocity, standard_velocity)

if __name__ == "__main__":
    main()
```
