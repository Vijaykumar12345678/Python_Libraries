```python
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Write a program to draw a swarm plot of total bill against size for a given dataset

```python
def strip_plot(df):
    """
    Description:
        Plots a strip plot to show the relationship between the party size and the total bill amount.

    Parameters:
        df (DataFrame): A pandas DataFrame containing the 'tips' dataset, which includes 'size' and 'total_bill' columns.

    Returns:
        None: This function does not return any value. It displays a strip plot.
    """
    sns.stripplot(x='size',y='total_bill',data=df,hue='size')
    plt.title('total bill against size ')
    plt.show()

def main():
    df=sns.load_dataset('tips')
    strip_plot(df)

if __name__=="__main__":
    main()
```

2. Write a program to draw swarm plot of “total bill” against day for a dataset given in url

```python
def swarm_plot(df):
    """
    Description:
        Plots a swarm plot to show the distribution of total bill amounts by day of the week.

    Parameters:
        df (DataFrame): A pandas DataFrame containing the 'tips' dataset, which includes 'day' and 'total_bill' columns.

    Returns:
        None: This function does not return any value. It displays a swarm plot.
   
    """
    sns.swarmplot(x='day',y='total_bill',data=df,hue='day')
    plt.title('total bill against day ')
    plt.show()

def main():
    df=sns.load_dataset('tips')
    
    swarm_plot(df)

if __name__=="__main__":
    main()
```
