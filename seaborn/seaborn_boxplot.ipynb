```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Write a program to draw box plot of life expectancy by continent for a data set given in a
url

```python
def boxplot(df):
    """
    Description:
        Plots a box plot to show the distribution of life expectancy by continent.

    Parameters:
        df (DataFrame): A pandas DataFrame containing data on various countries, including 'continent' and 'lifeExp' (life expectancy) columns.

    Returns:
        None: This function does not return any value. It displays a box plot.
    
    """
    sns.boxplot(x='continent',y='lifeExp',data=df,hue='continent')
    plt.title("life expectancy by continent")
    plt.show()

def main():
    df=pd.read_csv('https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv')
    boxplot(df)

if __name__=="__main__":
    main()
```

2. Write a program to draw a box plot of day by tips for a dataset given in a url

```python
def boxplot(df):
    
    """
    Description:
        Plots a box plot to show the distribution of tips by day of the week.

    Parameters:
        df (DataFrame): A pandas DataFrame containing the 'tips' dataset, which includes 'day' and 'tip' columns.

    Returns:
        None: This function does not return any value. It displays a box plot.
  
    
    """
    sns.boxplot(x='day',y='tip',data=df,hue='day')
    plt.title("day by tips")
    plt.show()

def main():
    df=sns.load_dataset('tips')
    
    boxplot(df)

if __name__=="__main__":
    main()
```
