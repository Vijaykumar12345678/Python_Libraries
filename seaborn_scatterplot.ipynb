```python
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url

```python
def scatter_plot(df):
    """
    Description:
        Plots a scatter plot showing the relationship between the day of the week and the total bill amount.

    Parameters:
        df (DataFrame): A pandas DataFrame containing the 'tips' dataset, which includes 'day' and 'total_bill' columns.

    Returns:
        None: This function does not return any value. It displays a scatter plot.
    """
    sns.scatterplot(x='day',y='total_bill',data=df,hue='day')
    plt.title("“day” against “total bill” ")
    plt.show()

def main():
    df=sns.load_dataset('tips')
    scatter_plot(df)

if __name__=="__main__":
    main()
```
