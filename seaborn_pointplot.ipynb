```python
import seaborn as sns

import matplotlib.pyplot as plt
```

1. Write a program to draw a point plot for sex against survived for a dataset given in url

```python
def point_plot(df):
    """
    Description:
        Plots a bar chart showing the survival rate of passengers on the Titanic based on gender.

    Parameters:
        df (DataFrame): A pandas DataFrame containing Titanic dataset information, including 'sex' and 'survived' columns.

    Returns:
        None: This function does not return any value. It displays a pointplot and prints the total number of survivors grouped by gender.

   
    """
    sns.pointplot(x="sex", y="survived", data=df)
    plt.title("sex against survived ")
    plt.show()
def main():
    df=sns.load_dataset('titanic')
    
    point_plot(df)
if __name__=="__main__":
    main()
```
